```python
import gymnasium as gym
import numpy as np
import matplotlib.pyplot as plt

# from wrappers import FireResetEnv
name = 'Pong-v5'
env = gym.make('ALE/'+name, render_mode="rgb_array")
env = env
env.reset()

# Get the number of actions and the action space
n_actions = env.action_space.n
print("Number of actions: ", n_actions)

# Get the observation space
obs, info = env.reset()
obs_shape = obs.shape
print("Observation space: ", obs_shape)
plt.imshow(obs)
plt.show()

# Fire the ball
obs, _, _, _, _ = env.step(5)
plt.imshow(obs)
plt.show()
```
